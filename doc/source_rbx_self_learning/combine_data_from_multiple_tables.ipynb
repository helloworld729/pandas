```python
import pandas as pd
```

#### 0 数据准备

```python
air_quality_no2 = pd.read_csv("../data/air_quality_no2_long.csv", parse_dates=True)
```

```python
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]
air_quality_no2.head()  # 时区 位置 参数1 指标1
```

```python
air_quality_pm25 = pd.read_csv("../data/air_quality_pm25_long.csv", parse_dates=True)
```

```python
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]
air_quality_pm25.head()  # 时区 位置 参数2 指标2
```

#### 1 conatenating

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)
air_quality.head()
```

```python
air_quality = air_quality.sort_values("date.utc")  # 按照时区排序
air_quality.head()
```

```python
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys=["PM25", "NO2"])
air_quality_.head()
```

#### 2 表连接

```python
stations_coord = pd.read_csv("../data/air_quality_stations.csv")  # 坐标
stations_coord.head()
```

```python
air_quality.head()
```

```python
print(air_quality.shape, stations_coord.shape, air_quality.shape)
```

```python
air_quality = pd.merge(air_quality, stations_coord, how="left", on="location")
air_quality.head()
```

```python
print(air_quality.shape, stations_coord.shape, air_quality.shape)
```

```python
air_quality_parameters = pd.read_csv("../data/air_quality_parameters.csv")
air_quality_parameters.head()
```

```python
air_quality = pd.merge(air_quality, air_quality_parameters,
                       how='left', left_on='parameter', right_on='id')
```
